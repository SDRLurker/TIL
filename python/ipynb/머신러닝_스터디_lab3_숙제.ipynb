```
import tensorflow as tf
x_data = [1, 2, 3]
y_data = [1, 2, 3]

#W = tf.Variable(tf.random_normal([1]), name='weight')
#b = tf.Variable(tf.random_normal([1]), name='bias')
W = tf.Variable(5.)
b = tf.Variable(4.)

X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

# 선형 모델 X * W에 대한 우리의 가설
# Our hypothesis for linear model X * W
hypothesis = X * W + b

# cost/loss 함수
# cost/loss function
cost = tf.reduce_mean(tf.square(hypothesis - Y))

# 최소화: 미분을 사용하여 경사하강
# Minimize: Gradient Descent using derivative
learning_rate = 0.1

gradientW = tf.reduce_mean((W * X + b - Y) * X) * 2
descentW = W - learning_rate * gradientW
updateW = W.assign(descentW)
gradientb = tf.reduce_mean((W * X + b - Y)) * 2
descentb = b - learning_rate * gradientb
updateb = b.assign(descentb)

# 세션에서 그래프를 시작합니다.
# Launch the graph in a session.
sess = tf.Session()
# 그래프에서 전역 변수들을 초기화합니다.
# Initiializes global variables in the graph.
sess.run(tf.global_variables_initializer())

for step in range(1000):
  if step % 20 == 0:
    print(step, sess.run(cost, feed_dict={X: x_data, Y: y_data}), sess.run(W), sess.run(b))
  sess.run([updateW, updateb], feed_dict={X: x_data, Y: y_data})
```

```
tf.reset_default_graph()

X = [1, 2, 3]
Y = [1, 2, 3]
# 잘못된 모델 weights를 설정
# Set wrong model weights
W = tf.Variable(5.)
b = tf.Variable(4.)
# 선형 모델 
# Linear model
hypothesis = X * W + b
# 수동 경사
# Manual gradient
gradientW = tf.reduce_mean((W * X + b - Y) * X) * 2
gradientb = tf.reduce_mean((W * X + b - Y)) * 2
# cost/loss 함수
# cost/loss function
cost = tf.reduce_mean(tf.square(hypothesis - Y))
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.1)

# 경사를 계산
# Get gradients
gvs = optimizer.compute_gradients(cost)
# 경사를 적용
# Apply gradients
apply_gradients = optimizer.apply_gradients(gvs)

# 세션에서 그래프를 시작합니다.
# Launch the graph in a session.
sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(1000):
  if step % 20 == 0:
    print(step, sess.run([cost, gradientW, gradientb, W, b, gvs]))
  else:
    sess.run([cost, gradientW, gradientb, W, b, gvs])
  sess.run(apply_gradients)
# step [gradientW, gradientb, W, b [(gvs의 gradientW, gvs의 W), (gvs의 gradientb, gvs의 b)]]
```
